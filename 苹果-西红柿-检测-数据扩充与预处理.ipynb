# 图像预处理

## 调整图像像素大小，生成灰度图并进行二值化处理

```python
import cv2

def preprocess_image(image_path, target_size=(200, 200)):
    # 读取图像
    image = cv2.imread(image_path)
    
    # 调整大小
    resized_image = cv2.resize(image, target_size)
    
    # 转换为灰度图像
    gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    
    # 二值化
    _, binary_image = cv2.threshold(gray_image, 125, 255, cv2.THRESH_BINARY)
    
    return binary_image

# 示例用法
image_path = "K:\\1900.png"
processed_image = preprocess_image(image_path)
cv2.imshow("Processed Image", processed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

def process_image(image_path):
    # 读取图像
    image = cv2.imread(image_path)
    
    # 转换为灰度图像
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # 进行图像二值化
    _, binary_image = cv2.threshold(gray_image, 120, 255, cv2.THRESH_BINARY)
    
    # 反转二值化图像
    inverted_image = cv2.bitwise_not(binary_image)
    
    return inverted_image

# 示例用法
image_path = "K:\\1900.png"
processed_image = process_image(image_path)
cv2.imshow("Processed Image", processed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 图片重命名脚本

```python
import os

def rename_images(folder_path, prefix="tomato"):
    # 获取文件夹中的所有文件名
    files = os.listdir(folder_path)
    
    # 统计图片文件的数量
    image_count = sum(1 for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')))
    
    # 遍历文件夹中的所有图片文件，并进行重命名
    for index, file in enumerate(files):
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            # 生成新的文件名
            new_filename = f"{prefix}_{index + 1:03d}.jpg"
            
            # 构建完整的文件路径
            old_filepath = os.path.join(folder_path, file)
            new_filepath = os.path.join(folder_path, new_filename)
            
            # 重命名文件
            os.rename(old_filepath, new_filepath)
            print(f"Renamed {file} to {new_filename}")

# 示例用法
folder_path = "S:\\apple_tomato_class\\test\\tomatoes"  # 替换为你的文件夹路径
rename_images(folder_path)
```

# 数据扩充脚本

```python
import os
from PIL import Image

def rotate_images(folder_path):
    # 获取文件夹中的所有文件名
    files = os.listdir(folder_path)
    
    # 创建保存旋转图像的文件夹
    output_folder = os.path.join(folder_path, "rotated_images")
    os.makedirs(output_folder, exist_ok=True)
    
    # 遍历文件夹中的所有图片文件，并进行旋转处理
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            # 构建完整的文件路径
            image_path = os.path.join(folder_path, file)
            
            # 打开图像文件
            with Image.open(image_path) as image:
                # 分别进行90度、180度、270度的旋转
                for angle in [90, 180, 270]:
                    rotated_image = image.rotate(angle, expand=True)
                    
                    # 构建保存旋转后图像的文件路径
                    output_path = os.path.join(output_folder, f"{os.path.splitext(file)[0]}_{angle}.jpg")
                    
                    # 保存旋转后的图像
                    rotated_image.save(output_path)
                    print(f"Rotated {file} by {angle} degrees and saved as {os.path.basename(output_path)}")

# 示例用法
folder_path = "S:\\apple_tomato_class\\train\\tomatoes"  # 替换为你的文件夹路径
rotate_images(folder_path)
```

# 统一图形像素大小

```python
import os
import cv2

def resize_images(folder_path, target_size=(248, 248)):
    # 获取文件夹中的所有文件名
    files = os.listdir(folder_path)
    
    # 创建保存调整大小后图像的文件夹
    output_folder = os.path.join(folder_path, "resized_images")
    os.makedirs(output_folder, exist_ok=True)
    
    # 遍历文件夹中的所有图片文件，并进行大小调整
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            # 构建完整的文件路径
            image_path = os.path.join(folder_path, file)
            
            # 读取图像
            image = cv2.imread(image_path)
            
            # 调整图像大小
            resized_image = cv2.resize(image, target_size)
            
            # 构建保存调整大小后图像的文件路径
            output_path = os.path.join(output_folder, file)
            
            # 保存调整大小后的图像
            cv2.imwrite(output_path, resized_image)
            print(f"Resized {file} and saved as {os.path.basename(output_path)}")

# 示例用法
folder_path = "S://apple_tomato_class//train/apples_rotated/"  # 替换为你的文件夹路径
resize_images(folder_path)
```

# 生成标签csv文件脚本

```python
import os
import csv

def generate_csv(folder_path, csv_filename):
    # 获取文件夹中的所有文件名
    files = os.listdir(folder_path)
    
    # 打开CSV文件，准备写入数据
    with open(csv_filename, 'w', newline='') as csvfile:
        fieldnames = ['Filename', 'Label']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        # 写入表头
        writer.writeheader()
        
        # 遍历文件夹中的所有文件，并写入CSV文件
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                # 获取文件名（不包含扩展名）
                filename = os.path.splitext(file)[0]
                
                # 从文件名中提取类别标签（这里假设类别信息从文件名中获取）
                if 'apple' in filename:
                    label = '0'
                elif 'tomato' in filename:
                    label = '1'
                else:
                    label = 'unknow'  # 如果无法从文件名中提取类别标签，则标记为'unknown'
                
                # 写入CSV文件
                writer.writerow({'Filename': file, 'Label': label})

# 示例用法
folder_path = "S:\\apple_tomato_class\\train\\mixed_data"  # 替换为你的文件夹路径
csv_filename = "S:\\apple_tomato_class\\train\\labels.csv"  # CSV文件名
generate_csv(folder_path, csv_filename)
```

# 训练二分类模型

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout

# 新的数据集路径
data_dir_train = "S:\\apple_tomato_class\\train"
data_dir_test = "S:\\apple_tomato_class\\test"

# 获取训练集和测试集中的类别列表
train_classes = os.listdir(data_dir_train)
test_classes = os.listdir(data_dir_test)

# 检查类别列表
print("训练集中的类别:", train_classes)
print("测试集中的类别:", test_classes)

# 映射类别标签到数字
label_to_index = dict((label, i) for i, label in enumerate(train_classes))
print("类别标签到数字的映射:", label_to_index)

# 反向映射数字到类别标签
index_to_label = dict((v, k) for k, v in label_to_index.items())
print("数字到类别标签的映射:", index_to_label)
```

```python
# 创建ImageDataGenerator实例，并进行数据增强和预处理
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # 80%用于训练，20%用于验证
)

# 使用ImageDataGenerator从DataFrame中加载图像数据，并进行批量处理
train_generator = datagen.flow_from_directory(
    directory=data_dir_train,
    target_size=(248, 248),
    batch_size=32,
    class_mode='binary',  # 二分类任务
    subset='training',  # 指定数据集为训练集
    shuffle=True  # 打乱数据集顺序
)

validation_generator = datagen.flow_from_directory(
    directory=data_dir_train,
    target_size=(248, 248),
    batch_size=32,
    class_mode='binary',  # 二分类任务
    subset='validation'  # 指定数据集为验证集
)

# 显示部分预处理结果
plt.figure(figsize=(10, 10))
for images, labels in train_generator:
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i])
        plt.title(index_to_label[int(labels[i])])
        plt.axis("off")
    break  # 仅显示一个批次的数据
plt.show()
```

```python
# 加载预训练的 ResNet50 模型，不包括顶层
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3), pooling='max')

# 创建顶层分类器
model = Sequential([
    base_model,
    Flatten(),
    Dropout(0.5),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])


# 冻结预训练模型的权重
base_model.trainable = False

# 编译模型
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # 二分类任务使用'binary_crossentropy'损失函数
              metrics=['accuracy'])
```

```python
# 训练模型
history = model.fit(train_generator, epochs=10, validation_data=validation_generator)

# 评估模型
test_loss, test_acc = model.evaluate(validation_generator)
```
